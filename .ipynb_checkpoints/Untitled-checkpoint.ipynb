```python
import pandas as pd

xls = pd.ExcelFile('MIDRC Open A1 and R1 - cumulative by batch.xlsx')
xls.sheet_names
```

```python
df = pd.read_excel(xls, 'Sex')
df
```

```python
pct_cols = [col for col in df.columns if '(%)' in col]
df[['date'] + pct_cols]
```

```python
xls_cdc = pd.ExcelFile('CDC_COVIDpos - cumulative by month.xlsx')
df_cdc = pd.read_excel(xls_cdc, 'Sex')
pct_cols_cdc = [col for col in df_cdc.columns if '(%)' in col]
df_cdc[['date'] + pct_cols]
```

```python
# Set the sheet name and the date
sheet_name = 'Sex'
date = np.datetime64('2022-12-31')

def calcjsd(sheet_name, date):
    #get the list of data columns for the specified sheet
    cols_to_use = midrc_data.sheets[sheet_name].data_columns
    
    #find the row with the highest date below the date specified above
    midrc_row = (np.searchsorted(midrc_data.sheets[sheet_name].df.date.values, date) - 1).clip(0)
    cdc_row = (np.searchsorted(cdc_data.sheets[sheet_name].df.date.values, date) - 1).clip(0)
    census_row = (np.searchsorted(census_data.sheets[sheet_name].df.date.values, date) - 1).clip(0)
    
    # get the data for the rows and columns determined above
    midrc_sheet_data  = np.asarray(midrc_data .sheets[sheet_name].df[cols_to_use].iloc[midrc_row].values,dtype=float)
    cdc_sheet_data    = np.asarray(cdc_data   .sheets[sheet_name].df[cols_to_use].iloc[cdc_row].values,dtype=float)
    census_sheet_data = np.asarray(census_data.sheets[sheet_name].df[cols_to_use].iloc[census_row].values,dtype=float)

    sheet_data = {'MIDRC':midrc_sheet_data, 'CDC':cdc_sheet_data, 'Census':census_sheet_data}
    distances = {'MIDRC vs CDC': distance.jensenshannon(midrc_sheet_data, cdc_sheet_data, base=2.0),
                 'MIDRC vs Census': distance.jensenshannon(midrc_sheet_data, census_sheet_data, base=2.0),
                 'CDC vs Census': distance.jensenshannon(cdc_sheet_data, census_sheet_data, base=2.0),
                }
    return distances, sheet_data

distances, sheet_data = calcjsd(sheet_name, date)

print(sheet_data['MIDRC'])
print(sheet_data['CDC'])
print(sheet_data['Census'])

print( f'midrc vs cdc JSD ({sheet_name}, {date}):    ', distances['MIDRC vs CDC'] )
print( f'midrc vs census JSD ({sheet_name}, {date}): ', distances['MIDRC vs Census'] )
print( f'cdc vs census JSD ({sheet_name}, {date}):   ', distances['CDC vs Census'] )
```

```python
# Set the sheet name and the date
# Options for sheet name: Sex, Race, Ethnicity, Race and Ethnicity
sheet_name = 'Race and Ethnicity'
date = np.datetime64('2022-12-31')

distances, sheet_data = calcjsd(sheet_name, date)

print(sheet_data['MIDRC'])
print(sheet_data['CDC'])
print(sheet_data['Census'])

print( f'midrc vs cdc JSD ({sheet_name}, {date}):    ', distances['MIDRC vs CDC'] )
print( f'midrc vs census JSD ({sheet_name}, {date}): ', distances['MIDRC vs Census'] )
print( f'cdc vs census JSD ({sheet_name}, {date}):   ', distances['CDC vs Census'] )
```
